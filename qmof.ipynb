### Shuffle Source CSV

```python
import pandas as pd

# Load the CSV file into a DataFrame
file_path = './labels/qmofruns_2.csv'
df = pd.read_csv(file_path, index_col = 0)

# Shuffle the rows of the DataFrame
shuffled_df = df.sample(frac=1).reset_index(drop=True)

# Save the shuffled DataFrame back to a CSV file
shuffled_df.to_csv('./labels/qmofruns_2_suffled.csv')
```

### Target Property - Oxo Formation Energy

#### - MOF level (for ALIGNN)

```python
df_qmof = pd.read_csv("./labels/qmofruns_2_suffled.csv", index_col = 0)
df_qmof['energy_sum'] = df_qmof['Oxo Formation Energy'] + df_qmof['Hydrogen Affinity Energy']
min_indxs = df_qmof.groupby('MOF')['energy_sum'].idxmin()

df_qmof = df_qmof.loc[min_indxs].drop(columns = ['energy_sum'])
df_qmof['sample'] = df_qmof.MOF + '.cif'
df_qmof[['sample', 'Oxo Formation Energy']].to_csv('./MOFs_oms/id_prop.csv', index=False, header=False)
```
